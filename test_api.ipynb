## Run this to Detect a Car

```python
!pip install requests
from PIL import Image
import matplotlib.pyplot as plt
import requests
import json

url = "http://127.0.0.1:8000/enhanced-vision"

with open("img123.jpg", "rb") as f:
    response = requests.post(url, files={"file": ("img123.jpg", f, "image/jpeg")})

print("Status Code:", response.status_code)
print("\nJSON Response:")
print(json.dumps(response.json(), indent=2))



img = Image.open("img123.jpg")
plt.imshow(img)
plt.axis("off")
plt.title("Carrrrr")
plt.show()
```

```python
data = response.json()

print("CNN Predictions:")
for label, score in data["cnn_predictions"].items():
    print(f"  {label}: {score:.2%}")

print("\nMultimodal Enhancement:")
print(data["multimodal_enhancement"])

print("\nFinal Enhanced Response:")
print(data["final_enhanced_response"])
```
